```python
import numpy as np
x = np.arange(0,100,1)
y = np.arange(0,100,1)
X = np.ones((100,100))
for i in range(100):
    for j in range(100):
        X[i,j]=x[i]+y[j]
        
print X
plt.imshow(X,plt.cm.Blues)
plt.show()
plt.imshow(X,plt.cm.hot)
plt.show()
```
